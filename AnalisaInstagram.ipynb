```python
#Langkah 2 - mengimport masing-masing library yang akan digunakan untuk memproses analisa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor

#tampilkan sekilas dataset yang akan diproses
data = pd.read_csv("Instagram.csv", encoding = 'latin1')
print(data.head())
```

```python
#selanjutnya pastikan jika ada data yang kosong atau (null)
data.isnull().sum()
```

```python
#hapus data-data yg null, tampilkan kembali data setelah didrop
data = data.dropna()
data.info()
```

```python
#selanjutnya mulai menganalisa jangkauan postingan instagram. Pertama cek distribusi impression yang didapatkan oleh user di seksi Home
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Impressions From Home")
sns.distplot(data['From Home'])
plt.show()
```

```python
#analisa jangkauang hastag
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Hashtags")
sns.distplot(data['From Hashtags'])
plt.show()
```

```python
#Analisa pula jangkuan dari seksi Explore
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Explore")
sns.distplot(data['From Explore'])
plt.show()
```

```python
#buat laporan persentasi impresi yang didapat dari variasi tadi
home = data["From Home"].sum()
hashtags = data["From Hashtags"].sum()
explore = data["From Explore"].sum()
other = data["From Other"].sum()

labels = ['From Home','From Hashtags','From Explore','Other']
values = [home, hashtags, explore, other]

fig = px.pie(data, values=values, names=labels, 
             title='Impressions on Instagram Posts From Various Sources', hole=0.5)
fig.show()
```

```python
#selanjutnya buat analisis konten dengan membuat sebuah visualisasi wordcloud dg menampilkan kata yang paling banyak digunakan untuk Caption
text = " ".join(i for i in data.Caption)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.style.use('classic')
plt.figure( figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show() 
```

```python
#buat pula visualisasi kata yang terbanyak dipakai untuk hastag
text = " ".join(i for i in data.Hashtags)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show() 
```

```python
#analisa pula relasi antara jumlah likes dengan jumlah impresi postingan
figure = px.scatter(data_frame = data, x="Impressions",
                    y="Likes", size="Likes", trendline="ols", 
                    title = "Relationship Between Likes and Impressions")
figure.show()
```

```python
#selain itu buat analisa antara jumlah komen dengan jumlah impresi
figure = px.scatter(data_frame = data, x="Impressions",
                    y="Comments", size="Comments", trendline="ols", 
                    title = "Relationship Between Comments and Total Impressions")
figure.show()
```

```python
#juga hubungan antara data yang dibagikan dengan impresi
figure = px.scatter(data_frame = data, x="Impressions",
                    y="Shares", size="Shares", trendline="ols", 
                    title = "Relationship Between Shares and Total Impressions")
figure.show()
```

```python
#selanjutnya buat analisa hubungan antara jumlah data disimpan dengan data impresi
figure = px.scatter(data_frame = data, x="Impressions",
                    y="Saves", size="Saves", trendline="ols", 
                    title = "Relationship Between Post Saves and Total Impressions")
figure.show()
```

```python
#dengan demikian, buat pula korelasi antara masing-masing kolom impresi
correlation = data.corr()
print(correlation["Impressions"].sort_values(ascending=False))
```

```python
#selanjutnya buat analisa rating konversi dimana terdapat rata-rata follower yang didapat dari jumlah kunjungan profil dari sebuah postingan
conversion_rate = (data["Follows"].sum() / data["Profile Visits"].sum()) * 100
print(conversion_rate)
```

```python
#buat analisa hubungan antara jumlah follower dengan total profil dikunjungi
figure = px.scatter(data_frame = data, x="Profile Visits",
                    y="Follows", size="Follows", trendline="ols", 
                    title = "Relationship Between Profile Visits and Followers Gained")
figure.show()
```

```python
#setelah itu, buatlah model yang melatih mesin untuk memprediksi jangkauan postingan instagram dengan metode ML. Pertama split data dahulu menjadi data latih dan data uji :
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 
                   'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42) 
```

```python
#setelah data displit, maka buat model 
model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

```python
#lakukan prediksi dengan melakukan input pada model
# Features = [['Likes','Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']]
features = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
model.predict(features)
```
